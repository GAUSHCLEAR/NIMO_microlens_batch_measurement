```python
import numpy as np
data_original = np.array([
    [21, 36],  
    [10, 38],  
    [36, 32],  
    [26, 37],  
    [36, 39],  
    [16, 35],  
    [28, 41]   
])
data_measured=data_original+np.random.normal(0,0.5,(7,2))
# 随机打乱data_measured
np.random.shuffle(data_measured)

# 定义旋转和平移
theta = np.pi / 5
R = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
t = np.array([1, 3])

# 应用变换
data_transformed = (data_original @ R.T) + t
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors



# 初始化NearestNeighbors对象
nbrs = NearestNeighbors(n_neighbors=1, algorithm='auto').fit(data_measured)
distances, indices = nbrs.kneighbors(data_transformed)

# 创建图形
plt.figure(figsize=(8, 6))
plt.scatter(data_measured[:, 0], data_measured[:, 1], color='blue', label='Measured Data')
plt.scatter(data_original[:, 0], data_original[:, 1], color='green', label='Original Data')
plt.scatter(data_transformed[:, 0], data_transformed[:, 1], color='red', label='Transformed Data')

# 绘制最近邻线
for i, index in enumerate(indices.flatten()):
    plt.plot([data_transformed[i, 0], data_measured[index, 0]], 
             [data_transformed[i, 1], data_measured[index, 1]], 'k--')

# 添加图例和标题
plt.legend()
plt.title('Data Points Visualization: Original, Transformed, and Measured')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.grid(True)
plt.show()
```
